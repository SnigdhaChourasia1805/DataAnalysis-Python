```python
# Import Pandas
```

```python
import pandas as pd
```

```python
# Load DataSet
```

```python
df=pd.read_csv("Countries_Data.csv")
print(df.head())
```

```python
# Which Columns Containing Missing Values
```

```python
df.isnull().sum()
```

```python
# Rows where political_leader is missing
```

```python
df[df["political_leader"].isnull()]
```

```python
# Rows where title is missing
```

```python
df[df["title"].isnull()]
```

```python
# replace missing values
```

```python
df["political_leader"]=df["political_leader"].fillna("Unknown")
```

```python
df["title"]=df["title"].fillna("Not Available")
```

```python
df.isnull().sum()
```

```python
# check data type of column
```

```python
df.dtypes
```

```python
# Convert Data
```

```python
df["population"]=df["population"].astype(int)
```

```python
# Column Names to LowerCase
```

```python
df.columns=df.columns.str.lower()
```

```python
# Rename column name
```

```python
df.rename(columns={"political_leader":"leader_name"},inplace=True)
```

```python
print(df.head())
```

```python
df.sort_values(by="population",ascending=False)
```

```python
df["country"].duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["population_density"]=df["population"]/df["urban_population"]
```
