```python
import numpy as np
df["Expense_Level"]=np.where(
    df["Total_Expense"]>40000,"High",
    np.where(df["Total_Expense"]>=25000,"Medium","Low")
)
```

```python
print(df[["Total_Expense","Expense_Level"]])
```

```python
df["Scholar_Status"]=np.where(
    df["Total_Expense"]<20000,
    "Eligible",
    "Not Eligible"
)

print(df[["Total_Expense","Scholarship_Status"]])
```
