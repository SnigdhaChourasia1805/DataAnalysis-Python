```python
df.groupby("region")["sales_revenue"].mean()
```

```python
df.groupby("region")["sales_revenue"].mean().plot(kind="bar")

plt.title("Average Sales Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Average Revenue")
plt.show()
```

```python
df.groupby("channel")["sales_revenue"].mean()
```

```python
sns.barplot(x="channel" , y="sales_revenue",data=df)

plt.title("Revenue by Marketing Channel")
plt.show()
```

```python
df.groupby("product_category")["sales_revenue"].mean()
```

```python
sns.barplot(x="product_category",y="sales_revenue",data=df)

plt.xticks(rotation=45)
plt.title("Revenue by Product Category")
plt.show()
```

```python
df.groupby("customer_segment")["sales_revenue"].mean()
```

```python
sns.barplot(x="customer_segment",y="sales_revenue",data=df)
plt.title("Revenue by Customer Segment")
plt.show()
```

```python
sns.histplot(df["sales_revenue"],bins=40,kde=True)
plt.title("Sales Revenue Distribution")
plt.show()
```

```python
sns.scatterplot(x="ad_spend",y="sales_revenue",data=df)
plt.title("Ad Spend VS Revenue")
plt.show()
```

```python
sns.scatterplot(x="price",y="sales_revenue",data=df)
plt.title("Price VS Revenue")
plt.show()
```

```python
sns.scatterplot(x="customer_lifetime_value",y="sales_revenue",data=df)
plt.title("Customer Life Time Value VS Revenue")
plt.show()
```

```python
corr=df.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True)
plt.title("Correlation MAtrix")
plt.show()
```
