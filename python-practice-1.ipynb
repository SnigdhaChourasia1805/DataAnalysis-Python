```python
# 9.Take a list of numbers and find the second largest number. 
```

```python
numbers = [70, 50, 40, 28]

largest = second = 0

for i in numbers:
    if i > largest:
        second = largest
        largest = i
    elif i > second and i != largest:
        second = i

print("Second largest:", second)
```

```python
# 11.Take a string and check whether it contains both letters and digits. 
```

```python
text = input("Enter a string: ")
has_letter = False
has_digit = False

for i in text:
    if i.isalpha():
        has_letter = True
    if i.isdigit():
        has_digit = True

if has_letter and has_digit:
    print("Contains letters and digits")
else:
    print("Does not contain both")
```

```python
# 12.Take a list of numbers and print the difference between maximum and minimum. 
```

```python
numbers = [60, 20, 40, 5]

maximum = minimum = numbers[0]

for i in numbers:
    if i > maximum:
        maximum = i
    if i < minimum:
        minimum = i

print("Difference:", maximum - minimum)
```

```python
# 15.Create a menu-driven program using while loop: 
# 1. Add numbers to a list 
# 2. Remove a number 
# 3. Display the list 
# 4. Exit 
```

```python
numbers = []

while True:
    print("1. Add number")
    print("2. Remove number")
    print("3. Display list")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        num = int(input("Enter number: "))
        numbers.append(num)

    elif choice == 2:
        num = int(input("Enter number to remove: "))
        if num in numbers:
            numbers.remove(num)

    elif choice == 3:
        print(numbers)

    elif choice == 4:
        break
```
